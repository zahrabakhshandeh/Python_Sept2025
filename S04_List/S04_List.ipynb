### Loop in List

```python
temperatures = [36.5, 36.5, 38.2, 39.3, 36.8]
#value, (value, index), index

#1: value
for i in temperatures:
    if i > 37:
        print(f"{i} -----> High")
```

```python
temperatures = [36.5, 36.5, 38.2, 39.3, 36.8]
i=0 
while i < len(temperatures):
    print(i, temperatures[i])
    i+=1
```

```python
temperatures = [36.5, 36.5, 38.2, 39.3, 36.8]
#value, (value, index), index

#1: Index ------> 0 - 4 (5), 0 -6 (7) -----> len
for i in range(len(temperatures)):
    if temperatures[i] > 37:
        print(f"patient_{i} --------> High")
```

```python
temperatures = [39.5, 36.5, 38.2, 39.3, 36.8]
#                 0     1     2     3     4
#value, index, (index, value)

#3: (index, value)
result = list(enumerate(temperatures))

for i, v in enumerate(temperatures):
    if v > 37:
        print(f"patient_{i+1}, {v} ------> High")
```

### Copy

```python
number1 = 56
number2 = number1 
number1 = 100 
print(number1, number2)
```

```python
#Assignment
temperatures = [36.5, 36.5, 38.2, 39.3, 36.8]
new_list = temperatures 
print(f"before update: list1 {temperatures}, list2 {new_list}")
temperatures.append(200)
print(f"after update: list1 {temperatures}, list2 {new_list}")
```

```python
# Shallow Copy
temperatures = [36.5, 36.5, 38.2]
new_list = temperatures[:]
print(f"before update: list1 {temperatures}, list2 {new_list}")
temperatures.append(200)
print(f"after update: list1 {temperatures}, list2 {new_list}")
```

```python
# Shallow Copy
temperatures = [36.5, 36.5, 38.2]
new_list = list(temperatures)
print(f"before update: list1 {temperatures}, list2 {new_list}")
temperatures.append(200)
print(f"after update: list1 {temperatures}, list2 {new_list}")
```

```python
# Shallow Copy
temperatures = [[36.5, 36.5], [38.2, 38]]
new_list = temperatures.copy()
print(f"before update: list1 {temperatures}, list2 {new_list}")
temperatures[0][0] = 40
print(f"after update: list1 {temperatures}, list2 {new_list}")
```

```python
# deep copy
import copy 
temperatures = [[36.5, 36.5], [38.2, 38]]
new_list = copy.deepcopy(temperatures)
print(f"before update: list1 {temperatures}, list2 {new_list}")
temperatures[0][0] = 40
print(f"after update: list1 {temperatures}, list2 {new_list}")
```

```python
temperatures = [40, 38.7, 36, 38.7, 38.7, 38.7]
max_tem = max(temperatures)
min_tem = min(temperatures)
sum_tem = sum(temperatures)
len_tem = len(temperatures)
avg = sum_tem / len_tem 
print(max_tem, min_tem, sum_tem, len_tem, avg)
```
