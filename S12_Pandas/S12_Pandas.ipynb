### Mapping

```python
data = {
    "City": ["New York", "London", "Chicago", "Manchester"],
    "Country": ["USA", "UK", "USA", "UK"]
}
df = pd.DataFrame(data)
mapping_data = {
    "USA": "United States",
    "UK": "Englan"
}
df["Country"] = df["Country"].map(mapping_data)
city_mapping = {
    "New York" : "NYC"
}
df["City"] = df["City"].replace(city_mapping)
df
```
