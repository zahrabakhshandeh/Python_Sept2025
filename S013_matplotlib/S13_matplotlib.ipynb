## Plot

```python
x = np.array([4, 8, 3, 10])
y = np.array([7, 8, 2, 5])
plt.figure(figsize=(4, 3))
plt.plot(x, y, label="XY", c="r", linewidth=1, linestyle=":", marker="o", ms=5)
plt.title("My Fist Plot", c="g", size=16)
plt.xlabel("X", size=16)
plt.ylabel("Y", size=16)
plt.legend()
plt.savefig("p1.png")
plt.show()
```

```python
x = np.linspace(-30, 30, 30)
# x, x*4, x**2
plt.figure(figsize=(4, 3))
plt.plot(x, x, label="x")
plt.plot(x, x*4, label="x*4")
plt.plot(x, x**2, label="x**2")
plt.legend()
plt.show()
```

## Scatter

```python
x = np.array([4, 8, 5, 7, 78, 3, 90, 80])
y = np.array([7, 8, 8, 9, 0, 2, 5, 5])
plt.figure(figsize=(4, 3))
plt.scatter(x, y, c="r", s=x)
plt.show()
```

```python
x1 = np.array([4, 8, 5, 7, 78, 3, 90, 80])
y1 = np.array([7, 8, 8, 9, 0, 2, 5, 5])

x2 = np.array([6, 8, 10])
y2 = np.array([7, 9, 23])
plt.figure(figsize=(4, 3))
plt.scatter(x1, y1)
plt.scatter(x2, y2, c=["black", "red", "green"])
plt.show()
```

## Pie Plot

```python
x = np.array([30, 20, 200])
y = np.array(["A", "B", "C"])
plt.figure(figsize=(4, 3))
plt.pie(x, labels=y, )
plt.legend()
plt.show()
```

```python
x = np.array([30, 25, 200])
y = np.array(["A", "B", "C"])
plt.figure(figsize=(4, 3))
plt.pie(x, labels=y, autopct="%1.2f%%")
plt.legend()
plt.show()
```

## Bar

```python
x = np.array(["A", "B", "C"])
y = np.array([30, 25, 200])
plt.figure(figsize=(4, 3))
plt.bar(x, y, width=0.4)
plt.legend()
plt.show()
```

```python
x = np.array(["A", "B", "C"])
y = np.array([30, 25, 200])
plt.figure(figsize=(4, 3))
plt.barh(x, y, height=0.4)
plt.legend()
plt.show()
```

## Subplot

```python
x = np.array([5, 10, 15, 20])
y = np.array([0, 4, 7, 10])

plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.plot(x, y)

plt.subplot(2, 3, 2)
plt.plot(x, x**2)

plt.subplot(2, 3, 3)
plt.pie(x)

plt.subplot(2, 3, 4)
plt.plot(x, y)

plt.subplot(2, 3, 5)
plt.plot(x, x**2)

plt.subplot(2, 3, 6)
plt.pie(x)



plt.show()
```

## 3D

```python
ax = plt.axes(projection="3d")
x = np.arange(0, 20, 0.2)
y = np.arange(40, 60, 0.2)
z = np.sin(x) * np.cos(x)
ax.plot(x, y, z)
plt.show()
```
