```python
dna = "AATGzCTTA"
a = t = g = c = 0
valid = True 

for i in dna:
    if i == "A":
        a += 1
    elif i == "T":
        t += 1
    elif i == "G":
        g += 1 
    elif i == "C":
        c += 1 
    else:
        valid = False
        break

if valid:
    print(f"A: {a}, T: {t}, G: {g}, C: {c}") 
else:
    print("Invalid DNA!")
```

```python
# A -------> T
# T -------> A
# G -------> C
# C -------> G 
# else ----> ?
dna = "AATGCTTA"
result = ""
for i in dna:
    if i == "A":
        result += "T"
    elif i == "T":
        result += "A"
    elif i == "G":
        result += "C"
    elif i == "C":
        result += "G"
    else:
        result += "?"

print(dna)
print(result)
```

```python
dna = "AATGCTTA"
result = dna.replace("A", "X")   #XXTGCTTX
result = result.replace("T", "A") #XXAGCAAX
result = result.replace("X", "T") #TTAGCAAT

result = result.replace("G", "Y") #TTAYCAAT
result = result.replace("C", "G") #TTAYGAAT
result = result.replace("Y", "C") #TTACGAAT
print(result) #TTACGAAT
```

### Continue

```python
n = int(input("Enter the number: ")) # 2
for i in range(n): #i=0, i=1
    weight1 = float(input("weight: ")) # 65
    height1 = float(input("height: ")) # 0
    if height1 == 0:
        print("Invalid!")
        continue

    bmi1 = weight1 / (height1**2) # 65 / 0
    print(f"BMI: {round(bmi1,2)}")
```
