#### float and int

###### 1.Arithmetic Operations

```python
# +, -, *, /, //, **, % 
total_tablets = 53 
dose_per_day = 4 
full_days = total_tablets // dose_per_day   #int
print(full_days)
```

```python
remaining = total_tablets % dose_per_day 
print(remaining)
```

```python
# += , -=, *=, /= 
weight = 56 
print(weight)
#weight = weight + 2
weight += 2
print(weight)
```

###### 2.Comparison Operations

```python
# <, >, <=, >=, ==, !=  ---------------> bool (True, False)
bmi = 30.1 
print("underweight: ", bmi < 18.5)
print("obses: ", bmi > 30)
```

```python
age = 50 
print(age == 60)
```

###### 3.Logical Operations

```python
# and
age = 65 
bmi = 32 
blood_pressure = 150

#            True           True          True
high_risk = age >= 65 and bmi > 30 and blood_pressure > 140 
print("Is the patient high risk? ", high_risk)
```

```python
# or 
bmi = 31
blood_pressure = 135 
#                                       False          False
#                                       True
print("high BMI or blood pressure: ", bmi > 30 or blood_pressure > 140)
```

```python
# not 
bmi = 22 
print("Not overweight: ", not (bmi > 25))
```
