```python
number1 = int(input("number1: ")) # 45
number2 = int(input("number2: ")) # 0
try:
    result_1 = number1 / number2  # result_1 = 2.0
except ZeroDivisionError as e:
    print(number1 * number2)
except NameError as e: 
    print(1)
except Exception as e:
    print(e)
else:
    print(result_1)
finally:
    print("end")
    
print(number1+number2)
```

```python
def function1(n1, n2):
    try:
        result = n1 / n2
        return result
    except ZeroDivisionError as e: 
        print(e)
        return -1
    else:
        print("else")
        print(result)
    finally:
        return 0
```

```python
result = function1(10, 5)
print(result)
```
