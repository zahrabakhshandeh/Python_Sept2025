### defult variables

```python
def sum_scores(math=0, python=0, english=0):
    print(f"Math:{math}, Python:{python}, English:{english}")
    sum_ = math + python + english 
    return sum_
```

```python
result = sum_scores(python=18)
print(result)
```

```python
def avg_scores(number, math=0, python=0, english=0):
    print(f"Math:{math}, Python:{python}, English:{english}")
    sum_ = math + python + english 
    avg = sum_ / number
    return avg
```

```python
avg = avg_scores(2, math=16, english=17)
print(avg)
```

```python
#map 
data = ["34.5", "36.7", "37.1", "38.4"]
new_data = []
for i in data:
    new_var = float(i)
    new_data.append(new_var)
print(new_data)
print(data)
```

```python
#map 
data = ["34.5", "36.7", "37.1", "38.4"]
new_list = list(map(float, data))
print(new_list)
```

```python
list1 = [23, 24, 12, 16, 45, 67]
list2 = [2, 56, 12, 17]
list3 = [23, 4, 10, 6]
new_list = []
for i in range(len(list1)):
    sum_numbers = list1[i]+list2[i]+list3[i]
    new_list.append(sum_numbers)
print(new_list)
```
